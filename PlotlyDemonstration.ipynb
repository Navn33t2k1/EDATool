```python
import plotly.offline as pyo
import plotly.graph_objs as go
import numpy as np
import pandas as pd
```

```python
ipl=delivery.merge(match, left_on='match_id', right_on='id')
```

```python
ipl.columns
```

```python
top50=ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(50).index.tolist()
```

```python
top50
```

```python
new_ipl=ipl[ipl['batsman'].isin(top50)]
```

```python
new_ipl
```

```python
new_ipl.shape
```

```python
# Calculating strik rate
runs=new_ipl.groupby('batsman')['batsman_runs'].sum()
```

```python
runs
```

```python
balls=new_ipl.groupby('batsman')['batsman_runs'].count()
```

```python
balls
```

```python
sr=(runs/balls)*100
```

```python
sr=sr.reset_index()
```

```python
sr.rename(columns={'batsman_runs':'SR'}, inplace=True)
sr.head()
```

```python
# Calculating avg.
# avg=(total number of runs)/(total number of outs)
# Calcutating number of outs for top 50 batsman
out=ipl[ipl['player_dismissed'].isin(top50)]
```

```python
out.head()
```

```python
nouts=out['player_dismissed'].value_counts()
```

```python
nouts
```

```python
avg=runs/nouts
avg.head()
```

```python
avg=avg.reset_index()
avg.head()
```

```python
avg.rename(columns={'index':'batsman',0:'avg'}, inplace=True)
```

```python
avg.head()
```

```python
avg=avg.merge(sr, on='batsman')
```

```python
avg.head()
```

```python
#trace=go.Scatter(x=avg['avg'], y=avg['SR'], mode='markers')
# Adding text to point
#trace=go.Scatter(x=avg['avg'], y=avg['SR'], mode='markers',text=avg['batsman'])
# Adding color to point
#trace=go.Scatter(x=avg['avg'], y=avg['SR'], mode='markers',text=avg['batsman'], marker={'color':'#00a65a'})
# Increasing size
trace=go.Scatter(x=avg['avg'], y=avg['SR'], mode='markers',text=avg['batsman'], marker={'color':'#00a65a', 'size':16})
data=[trace]
layout=go.Layout(title='Batsman Avg vs SR', xaxis={'title':'Batsman Average'}, yaxis={'title':'Batsman Strike Rate'})
fig=go.Figure(data=data, layout=layout)
pyo.plot(fig, filename='myfile.html')
```

```python
# Year by year batsman Performance
single=ipl[ipl['batsman']=='V Kohli']
performance=single.groupby('Season')['batsman_runs'].sum().reset_index()


single1=ipl[ipl['batsman']=='MS Dhoni']
performance1=single1.groupby('Season')['batsman_runs'].sum().reset_index()
performance1
```

```python
performance.rename(columns={'batsman_runs':'total_runs'}, inplace=True)
performance1.rename(columns={'batsman_runs':'total_runs'}, inplace=True)
```

```python
performance
```

```python
# Plot lines
#trace=go.Scatter(x=performance['Season'], y=performance['total_runs'], mode='lines', marker={'color':'#00a65a'})
trace=go.Scatter(x=performance['Season'], y=performance['total_runs'], mode='lines+markers', marker={'color':'#00a65a'}, name='V Kohli')
trace1=go.Scatter(x=performance1['Season'], y=performance1['total_runs'], mode='lines+markers', name='MS Dhoni')
data=[trace, trace1]
layout=go.Layout(title='Year by year performance',
                xaxis={'title':'Season'},
                yaxis={'title':'Total_runs'})
fig=go.Figure(data=data, layout=layout)
pyo.plot(fig)
```

```python
# Multiple lines charts

def batsman_comp(*name):
    data=[]
    for i in name:
        single=ipl[ipl['batsman']==i]
        performance=single.groupby('Season')['batsman_runs'].sum().reset_index()
        trace=go.Scatter(x=performance['Season'], y=performance['batsman_runs'], mode='lines+markers',
                        name=i)
        data.append(trace)
    layout=go.Layout(title='Batsman records comparators', xaxis={'title':'Season'}, yaxis={'title':'Runs'})
    fig=go.Figure(data=data, layout=layout)
    pyo.plot(fig, filename='year_by_year')
```

```python
batsman_comp('V Kohli', 'RG Sharma', 'DA Warner', 'MS Dhoni')
```

```python
# there are 3 types of bar graphs

iw=top10_df.groupby(['batsman', 'inning'])['batsman_runs'].sum().reset_index()
mask1=iw['inning']==1
mask2=iw['inning']==2

one=iw[mask1]
two=iw[mask2]
```

```python
one.rename(columns={'batsman_runs':'1innings'}, inplace=True)
two.rename(columns={'batsman_runs':'2innings'}, inplace=True)
final=one.merge(two, on='batsman')[['batsman', '1innings', '2innings']]
final
```

```python
trace1=go.Bar(x=final['batsman'], y=final['1innings'], name='1innings', marker={'color':'#00a65a'})
trace2=go.Bar(x=final['batsman'], y=final['2innings'], name='2innings', marker={'color':'#a6a65a'})
data=[trace1, trace2]
layout=go.Layout(title='Inning wise Score', xaxis={'title':'Bastsman'}, yaxis={'title':'Runs'})
fig=go.Figure(data=data, layout=layout)
pyo.plot(fig)
```

```python
# Plot Heatman
trace=go.Heatmap(x=six['batting_team'], y=six['over'], z=six['count(six)'])
data=[trace]
layout=go.Layout(title='Six Heatman')
fig=go.Figure(data=data, layout=layout)
pyo.plot(fig)
```

```python
dots=delivery[delivery['batsman_runs']==0]
dots=dots.groupby(['batting_team', 'over'])['batsman_runs'].count().reset_index()
dots.rename(columns={'batsman_runs':'count(dots)'}, inplace=True)

from plotly import tools

trace1=go.Heatmap(x=six['batting_team'], y=six['over'], z=six['count(six)'].values.tolist())
trace2=go.Heatmap(x=dots['batting_team'], y=dots['over'], z=dots['count(dots)'].values.tolist())
fig=tools.make_subplots(rows=1, cols=2, subplot_titles=["6's,", "0's"], shared_yaxes=True )
fig.append_trace(trace1, 1,1)
fig.append_trace(trace2, 1,2)
pyo.plot(fig)
```
